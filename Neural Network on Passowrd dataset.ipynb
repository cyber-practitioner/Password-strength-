```python
def passwords(url):
    return url
```

```python
def getLength(url):
  length=len(str(url))          
  return length
```

```python
def haveAtSign(url):
    AT=str(url)

    return AT.count('@')
def haveexclamation(url):
    return str(url).count("!")
```

```python
def havehash(url):
    return str(url).count("#")
def havepercentage(url):
    return str(url).count("%")
def havedollar(url):
    return str(url).count("$")
def haveand(url):
    return str(url).count("&")
def havecolon(url):
    return str(url).count(';')
```

```python
def Count(sr):
    upper = 0
    s=str(sr)
    for i in range(0,len(s)):
        if s[i].isupper():
            upper += 1
    return upper

```

```python
def lcount(url):
    lower= 0
    s=str(url)
    for i in range(0,len(s)):
        if s[i].islower():
            lower += 1
    return lower
```

```python

def digits(url):
    digit=0
    s=str(url)
    for i in s:
        if i.isdigit():
            digit += 1
    return digit
```

##### FEATURE SET 

```python
def featureExtraction(url,label):
#11 features
    features = []
    features.append(passwords(url))
    features.append(getLength(url))
    features.append(haveAtSign(url))
    features.append(haveexclamation(url))
    features.append(havehash(url))
    features.append(havepercentage(url))
    features.append(havedollar(url))
    features.append(haveand(url))
    features.append(havecolon(url))
    features.append(Count(url))
    features.append(lcount(url))
    features.append(digits(url))
    #features.append(final(url))

    return features
```

```python
legi_features = []
label=0

for i in range(0,669640):
  url = legiurl['password'][i]
  legi_features.append(featureExtraction(url,label))
```
